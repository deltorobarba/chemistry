#### **Electronic Energy Levels of hydrogen-like atom**

Example that demonstrates spectral analysis for a single atom by calculating its electronic energy levels. This example uses QuTip, a Python package for quantum optics simulations, to model the energy spectrum of a single atom such as alkaline earth metals or halogens.

We can compute the energy levels of a hydrogen-like atom using a simplified quantum mechanical model. This method can be adapted for atoms like alkaline earth metals or halogens by considering their electronic structure and solving for electronic transitions.


Electronic Energy Levels of a Hydrogen-Like Atom


```python
!pip install qutip -q
import numpy as np
import matplotlib.pyplot as plt
from qutip import basis, qeye, tensor, destroy

# Constants (in atomic units)
Z = 1  # Atomic number (Z = 1 for hydrogen, Z = 2 for helium, etc.)
n_max = 10  # Maximum principal quantum number

# Define the Hamiltonian for a hydrogen-like atom
def hydrogen_like_hamiltonian(Z, n_max):
    energies = []
    for n in range(1, n_max + 1):
        # Energy levels follow the formula: E_n = -Z^2 / (2 * n^2)
        energy_n = -Z**2 / (2 * n**2)
        energies.append(energy_n)
    return np.array(energies)

# Calculate the energy levels
energy_levels = hydrogen_like_hamiltonian(Z, n_max)

# Plot the energy levels (as lines)
plt.figure(figsize=(6, 4))
for n, energy in enumerate(energy_levels, start=1):
    plt.hlines(energy, xmin=n - 0.5, xmax=n + 0.5, color='blue', lw=2)
    plt.text(n, energy + 0.005, f'n={n}', ha='center')

plt.xlabel('Principal Quantum Number (n)')
plt.ylabel('Energy (a.u.)')
plt.title('Electronic Energy Levels of Hydrogen-Like Atom (Z=1)')
plt.grid(True)
plt.show()

# Print the energy levels
print("Energy levels (in atomic units):")
for n, energy in enumerate(energy_levels, start=1):
    print(f"n={n}, Energy: {energy:.6f} a.u.")
```

1. **Quantum Mechanical Model**: We use a simplified quantum mechanical model to calculate the energy levels of a hydrogen-like atom. The energy levels follow the well-known formula $ E_n = -\frac{Z^2}{2n^2} $, where \( Z \) is the atomic number and \( n \) is the principal quantum number.
2. **Hamiltonian**: In this case, the Hamiltonian is modeled as a diagonal matrix of the energy levels for different values of \( n \), where \( n \) represents different quantum states.
3. **Visualization**: The energy levels are plotted as horizontal lines, with each line corresponding to a principal quantum number \( n \).

**Adaptation for Different Atoms:**

- This model, however, does not account for multi-electron effects, spin-orbit coupling, or fine structure. To include these, more advanced quantum chemistry methods are needed (e.g., solving the Schrödinger equation with appropriate potentials for many-electron systems).

**Output**

- The plot will show the energy levels for different quantum numbers \( n \) (1 through 10, by default).
- The energy levels will be printed in **atomic units** (Hartrees).

For hydrogen, the electronic transitions between energy levels are well-known.

The corresponding emission and absorption spectra can be calculated using the Rydberg formula:

$
\frac{1}{\lambda} = R_H \left( \frac{1}{n_1^2} - \frac{1}{n_2^2} \right)
$

Where:
- $ \lambda $ is the wavelength of the transition.
- $ R_H $ is the Rydberg constant for hydrogen, $ R_H = 1.097 \times 10^7 \, \text{m}^{-1} $.
- $ n_1 $ and $ n_2 $ are the principal quantum numbers of the lower and higher energy levels, respectively, with $ n_2 > n_1 $.

For hydrogen, the transitions typically seen in the visible spectrum come from the Balmer series ($ n_1 = 2 $).


```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
R_H = 1.097e7  # Rydberg constant in m^-1
c = 3e8  # Speed of light in m/s
h = 6.626e-34  # Planck's constant in J·s
eV_to_J = 1.60218e-19  # Conversion from eV to Joules

# Rydberg formula for hydrogen transitions (in nm)
def rydberg_wavelength(n1, n2):
    return 1e9 / (R_H * (1/n1**2 - 1/n2**2))  # Wavelength in nm

# Step 1: Define the possible transitions (Balmer series for visible light)
transitions = []
for n2 in range(3, 7):  # Transitions to n1=2 from n2=3, 4, 5, 6
    transitions.append((2, n2))

# Step 2: Calculate the wavelengths for these transitions
wavelengths = [rydberg_wavelength(n1, n2) for n1, n2 in transitions]

# Step 3: Simulate and plot the emission and absorption spectra
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Emission Spectrum (peaks at specific wavelengths)
axs[0].bar(wavelengths, [1.0] * len(wavelengths), width=20.0, color='orange')
axs[0].set_xlabel('Wavelength (nm)')
axs[0].set_ylabel('Intensity (a.u.)')
axs[0].set_title('Emission Spectrum of Hydrogen (Balmer Series)')
axs[0].grid(True)

# Absorption Spectrum (dips at specific wavelengths)
axs[1].plot(np.linspace(350, 700, 1000), np.ones(1000), color='gray')  # Baseline at 1.0 intensity
for wavelength in wavelengths:
    axs[1].plot([wavelength, wavelength], [1.0, 0.0], color='blue', lw=2)  # Vertical absorption dips
axs[1].set_xlabel('Wavelength (nm)')
axs[1].set_ylabel('Normalized Intensity (a.u.)')
axs[1].set_title('Absorption Spectrum of Hydrogen (Balmer Series)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Step 4: Print out the transition wavelengths
print("Balmer series transitions (wavelengths in nm):")
for (n1, n2), wavelength in zip(transitions, wavelengths):
    print(f"n2={n2} -> n1={n1}: {wavelength:.2f} nm")
```

- **Rydberg Formula:** This formula calculates the wavelength of light emitted or absorbed when an electron transitions between energy levels.
- **Balmer Series:** We focus on transitions where the electron ends up at $ n_1 = 2 $ (visible light region). These include $ n_2 = 3, 4, 5, 6 $, which correspond to specific emission and absorption lines in the visible range.
- **Wavelengths:** We calculate the wavelengths of the transitions and plot the emission (peaks) and absorption (dips) spectra.

**Output:**
- **Emission Spectrum:** Shows peaks at specific wavelengths corresponding to hydrogen’s visible spectral lines.
- **Absorption Spectrum:** Shows dips at the same wavelengths, corresponding to the absorption of photons at those energies.
- **Wavelengths:** These are the wavelengths corresponding to the visible Balmer series transitions.

For hydrogen's Balmer series, the most important lines are around:

- $ n_2 = 3 \rightarrow n_1 = 2 $: 656.3 nm (H-alpha)
- $ n_2 = 4 \rightarrow n_1 = 2 $: 486.1 nm (H-beta)
- $ n_2 = 5 \rightarrow n_1 = 2 $: 434.0 nm (H-gamma)
- $ n_2 = 6 \rightarrow n_1 = 2 $: 410.2 nm (H-delta)

Here’s how you convert the energy difference (in atomic units) to a wavelength:

**Step 1: Energy-Wavelength Relationship**
The energy \( E \) of a photon is related to its wavelength \( \lambda \) by the following equation:

\[
E = \frac{hc}{\lambda}
\]

Where:
- \( E \) is the energy of the photon (in Joules),
- \( h \) is Planck's constant \( (6.626 \times 10^{-34} \, \text{J·s}) \),
- \( c \) is the speed of light \( (3.0 \times 10^8 \, \text{m/s}) \),
- \( \lambda \) is the wavelength of the photon (in meters).

**Step 2: Converting Atomic Units to Joules**
In atomic units (a.u.), energy is expressed in terms of the Hartree energy. The Hartree energy \( E_h \) is approximately:

\[
1 \, \text{a.u.} = 4.3597 \times 10^{-18} \, \text{J}
\]

So, if the energy difference between \( n=3 \) and \( n=2 \) is \( 0.069444 \, \text{a.u.} \), you can convert this to Joules as follows:

\[
E_{\text{photon}} = 0.069444 \, \text{a.u.} \times 4.3597 \times 10^{-18} \, \text{J/a.u.} = 3.027 \times 10^{-19} \, \text{J}
\]

**Step 3: Solving for Wavelength**
Now, we can rearrange the energy-wavelength equation to solve for \( \lambda \):

\[
\lambda = \frac{hc}{E}
\]

Substitute \( h = 6.626 \times 10^{-34} \, \text{J·s} \), \( c = 3.0 \times 10^8 \, \text{m/s} \), and \( E = 3.027 \times 10^{-19} \, \text{J} \):

\[
\lambda = \frac{6.626 \times 10^{-34} \times 3.0 \times 10^8}{3.027 \times 10^{-19}} = 6.565 \times 10^{-7} \, \text{m}
\]

This gives:

\[
\lambda = 656.5 \, \text{nm}
\]

**Conclusion**
The energy difference between the \( n=3 \) and \( n=2 \) levels of hydrogen, \( 0.069444 \, \text{a.u.} \), corresponds to a photon with a wavelength of approximately **656.5 nm**. This matches very closely with the experimental value of 656.34 nm, which is the red spectral line known as the **H-alpha line** in the Balmer series.

So, to summarize:
- The difference in energy \( 0.069444 \, \text{a.u.} \) is the energy released when an electron transitions from \( n=3 \) to \( n=2 \).
- This energy corresponds to a photon with a wavelength of about **656.5 nm**.

#### **Electronic Energy Levels of Calcium (Ca)**

**Let's extend the previous example for an alkaline earth metal, such as calcium (Ca). Alkaline earth metals (like magnesium, calcium, etc.) have two electrons in their outermost shell, making their electronic structure more complex than hydrogen-like atoms. To approximate their energy levels, we can use quantum defect theory.**

**Quantum Defect Theory**

In multi-electron systems like calcium, the inner electrons shield the nucleus, leading to a slight modification of the energy levels compared to hydrogen-like atoms. The energy levels are given by the formula:

$
E_n = -\frac{Z_{\text{eff}}^2}{2(n - \delta_l)^2}
$

Where:
- $ Z_{\text{eff}} $ is the effective nuclear charge (the positive charge felt by the outermost electron after accounting for inner electron shielding).
- $ n $ is the principal quantum number.
- $ \delta_l $ is the quantum defect, which accounts for deviations from the hydrogen-like energy levels due to electron-electron interactions and the penetration of the electron into inner shells.

**Constants for Calcium (Ca):**
- Atomic number $ Z = 20 $
- Effective nuclear charge for the outer electron $ Z_{\text{eff}} \approx 2 $ (rough approximation)
- Quantum defect $ \delta_s = 1.7 $, $ \delta_p = 0.3 $, $ \delta_d = 0.05 $

```python
!pip install qutip -q
import numpy as np
import matplotlib.pyplot as plt

# Constants for Calcium (Ca)
Z_eff = 2  # Approximate effective nuclear charge for calcium
n_max = 10  # Maximum principal quantum number
quantum_defects = {'s': 1.7, 'p': 0.3, 'd': 0.05}  # Quantum defects for different orbitals

# Define the Hamiltonian for an alkaline earth metal (e.g., calcium)
def alkaline_earth_hamiltonian(Z_eff, n_max, quantum_defect):
    energies = []
    for n in range(1, n_max + 1):
        if n - quantum_defect > 0:  # Ensure we don't calculate negative n-quantum_defect
            energy_n = -Z_eff**2 / (2 * (n - quantum_defect)**2)
            energies.append(energy_n)
    return np.array(energies)

# Calculate the energy levels for different orbitals (s, p, d)
energy_levels_s = alkaline_earth_hamiltonian(Z_eff, n_max, quantum_defects['s'])
energy_levels_p = alkaline_earth_hamiltonian(Z_eff, n_max, quantum_defects['p'])
energy_levels_d = alkaline_earth_hamiltonian(Z_eff, n_max, quantum_defects['d'])

# Plot the energy levels (as lines)
plt.figure(figsize=(12, 6))  # Adjusted the figure size to make it wider

# Plot 's' orbital energy levels
for n, energy in enumerate(energy_levels_s, start=1):
    plt.hlines(energy, xmin=n - 0.25, xmax=n + 0.25, color='blue', lw=2)
    plt.text(n, energy + 0.005, f'n={n} (s)', ha='center', fontsize=10)

# Plot 'p' orbital energy levels
for n, energy in enumerate(energy_levels_p, start=1):
    plt.hlines(energy, xmin=n - 0.25, xmax=n + 0.25, color='red', lw=2)
    plt.text(n, energy + 0.005, f'n={n} (p)', ha='center', fontsize=10)

# Plot 'd' orbital energy levels
for n, energy in enumerate(energy_levels_d, start=1):
    plt.hlines(energy, xmin=n - 0.25, xmax=n + 0.25, color='green', lw=2)
    plt.text(n, energy + 0.005, f'n={n} (d)', ha='center', fontsize=10)

# Adjust the x-axis range and add space between the labels
plt.xlim(0, n_max + 2)  # Extend the x-axis a bit beyond n_max
plt.xlabel('Principal Quantum Number (n)')
plt.ylabel('Energy (a.u.)')
plt.title('Electronic Energy Levels of Calcium (Ca)')
plt.grid(True)
plt.show()

# Print energy levels
print("Energy levels (in atomic units):")
for n, (energy_s, energy_p, energy_d) in enumerate(zip(energy_levels_s, energy_levels_p, energy_levels_d), start=1):
    print(f"n={n}, s: {energy_s:.6f} a.u., p: {energy_p:.6f} a.u., d: {energy_d:.6f} a.u.")
```

1. **Quantum Defect Theory**: This model uses quantum defects to adjust the energy levels for each orbital type:
   - $ s $-orbital quantum defect $ \delta_s = 1.7 $
   - $ p $-orbital quantum defect $ \delta_p = 0.3 $
   - $ d $-orbital quantum defect $ \delta_d = 0.05 $
   
   The energy levels are adjusted accordingly for each orbital.

2. **Plotting**: The code plots the energy levels for the **s**, **p**, and **d** orbitals as separate lines with different colors.

3. **Effective Nuclear Charge**: We use an approximate effective nuclear charge $ Z_{\text{eff}} = 2 $ for calcium, which accounts for the shielding effects from inner electrons.

**Output:**

- A plot showing the **electronic energy levels** for the **s**, **p**, and **d** orbitals for calcium.
- The energy levels are printed in **atomic units** (Hartrees).

**Adapting to Other Alkaline Earth Metals:**
For other alkaline earth metals (like **magnesium**, **strontium**, etc.), you can adjust:
- The **atomic number** $ Z $
- The **effective nuclear charge** $ Z_{\text{eff}} $ based on shielding.
- The **quantum defect** values specific to the element.

This approach provides an approximation of the energy spectrum, and for more detailed results, a full quantum mechanical solution with electron-electron interactions would be necessary.

#### **Emission and Absorption Spectrum of Calcium (Ca)**

For the emission and absorption spectra of an atom like calcium, is different from the molecular case, since this is about electronic transitions instead of vibrational modes.

Calcium has notable spectral lines, especially the ones corresponding to transitions between the 4s, 3d, and 4p orbitals.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the energy levels of calcium
# These are simplified and approximate values for major transitions in eV
energy_levels = {
    '4s': 0.0,  # Ground state (4s)
    '4p': 2.93, # First excited state (4p)
    '3d': 1.88  # Second excited state (3d)
}

# Step 2: Define possible transitions (in eV)
transitions = {
    ('4p', '4s'): energy_levels['4p'] - energy_levels['4s'],
    ('3d', '4s'): energy_levels['3d'] - energy_levels['4s'],
    ('4p', '3d'): energy_levels['4p'] - energy_levels['3d']
}

# Step 3: Convert transition energies from eV to wavelength in nm (using E = hc/λ)
def energy_to_wavelength(energy_eV):
    h = 4.1357e-15  # Planck's constant in eV·s
    c = 3e8  # Speed of light in m/s
    return (h * c) / energy_eV * 1e9  # Convert to nm

# Calculate wavelengths of transitions
wavelengths = {transition: energy_to_wavelength(energy) for transition, energy in transitions.items()}

# Step 4: Simulate and plot the emission and absorption spectra
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Emission Spectrum (peaks at specific wavelengths)
axs[0].bar(wavelengths.values(), [1.0] * len(wavelengths), width=20.0, color='orange')
axs[0].set_xlabel('Wavelength (nm)')
axs[0].set_ylabel('Intensity (a.u.)')
axs[0].set_title('Emission Spectrum of Calcium (Electronic Transitions)')
axs[0].grid(True)

# Absorption Spectrum (dips at specific wavelengths)
axs[1].plot(np.linspace(200, 1000, 1000), np.ones(1000), color='gray')  # Baseline at 1.0 intensity
for wavelength in wavelengths.values():
    axs[1].plot([wavelength, wavelength], [1.0, 0.0], color='blue', lw=2)  # Vertical absorption dips
axs[1].set_xlabel('Wavelength (nm)')
axs[1].set_ylabel('Normalized Intensity (a.u.)')
axs[1].set_title('Absorption Spectrum of Calcium (Electronic Transitions)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Step 5: Print out the transition wavelengths
print("Electronic transitions (wavelengths in nm):")
for transition, wavelength in wavelengths.items():
    print(f"{transition[0]} -> {transition[1]}: {wavelength:.2f} nm")
```

- **Energy Levels:** We are using simplified energy values for the 4s, 4p, and 3d orbitals of calcium.
- **Transitions:** We calculate the possible transitions between these states and their corresponding energies.
- **Wavelengths:** The energy of each transition is converted into a wavelength (in nm) using $ E = \frac{hc}{\lambda} $
- **Spectra:** The emission spectrum shows peaks where light is emitted, while the absorption spectrum shows dips where light is absorbed.
* Emission spectrum (peaks at specific wavelengths corresponding to transitions). Absorption spectrum (dips at the same wavelengths). Printed list of the wavelengths corresponding to the transitions.
